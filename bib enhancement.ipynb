```python
def initial(s):
    out = ''
    for i, c in enumerate(s):
        if c.isupper():
            if c == 'Y' and s[i+1] == 'u':
                c = 'Yu'
            out += c +'. '

        if c == '-':
            out = out[:-1] + c 
    return out

with open('lib.bib') as bibtex_file:
    parser = BibTexParser()
#     parser.customization = customizations
    bib_database = bibtexparser.load(bibtex_file, parser=parser)
    
    
    for entry in bib_database.entries:
        print(entry['author'])
        auth_list = entry['author'].split(' and ')
        
        ini_auth_list = ''
        for auth in auth_list:
#             print(auth)
            if auth != 'others':
                last, first = auth.split(', ')
                ini_name = last + ', ' + initial(first)
            
            ini_auth_list += ini_name + 'and '
            
        entry['author'] = ini_auth_list[:-4]
        
        print(entry['author'])
        
        
with open('lib_new.bib', 'w') as bibtex_file:
    bibtexparser.dump(bib_database, bibtex_file)                
                
```
